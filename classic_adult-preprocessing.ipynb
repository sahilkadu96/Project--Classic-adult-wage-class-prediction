```python
import numpy as np
import pandas as pd
```

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header = None)
train_set
```

```python
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',  'marital_status', 'occupation', 'relationship', 
              'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'wage_class']
train_set.columns = col_labels
train_set
```

```python
train_set['workclass'].replace({' ?': 'NA wc'}, inplace = True)
train_set['workclass'] = train_set['workclass'].apply(lambda x: x.strip())
train_set['workclass'].unique()
```

```python
workclass_dummy = pd.get_dummies(train_set['workclass'])
workclass_dummy
```

```python
train_set['education'].unique()
```

```python
train_set['education_num'].unique()
```

```python
train_set['marital_status'] = train_set['marital_status'].apply(lambda x: x.strip())
train_set['marital_status'].unique()
```

```python
marital_status_dummy = pd.get_dummies(train_set['marital_status'])
marital_status_dummy
```

```python
train_set['occupation'].replace({' ?': 'NA o'}, inplace = True)
train_set['occupation'] = train_set['occupation'].apply(lambda x: x.strip())
train_set['occupation'].unique()
```

```python
occupation_dummy = pd.get_dummies(train_set['occupation'])
occupation_dummy
```

```python
train_set['relationship'] = train_set['relationship'].apply(lambda x: x.strip())
train_set['relationship'].unique()
```

```python
relationship_dummy = pd.get_dummies(train_set['relationship'])
relationship_dummy
```

```python
train_set['race'] = train_set['race'].apply(lambda x: x.strip())
train_set['race'].unique()
```

```python
race_dummy = pd.get_dummies(train_set['race'])
race_dummy
```

```python
train_set['sex'].replace({' Female':0, ' Male':1}, inplace = True)
train_set['sex'].unique()
```

```python
train_set['native_country'].replace({'?': 'NA nc'}, inplace = True)
train_set['native_country'] = train_set['native_country'].apply(lambda x: x.strip())
train_set['native_country'].unique()
```

```python
native_country_dummy = pd.get_dummies(train_set['native_country'])
native_country_dummy
```

```python
dummies = pd.concat([workclass_dummy, marital_status_dummy, occupation_dummy, relationship_dummy,
                    race_dummy, native_country_dummy], axis = 1)
dummies
```

```python
dummies.columns
```

```python
columns_drop = ['workclass', 'education', 'marital_status', 'occupation', 'relationship', 'race', 'native_country']

train_set.drop(columns_drop, axis = 1, inplace = True)
train_set
```

```python
final_data = pd.concat([dummies, train_set], axis = 1)
final_data
```

```python
final_data.to_csv('classic_adult_preproc.csv', index = False)
```
